```python
import pandas as pd
```

```python
d=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
d
```

```python
x=d.drop(columns=['Gender','Purchased','User ID'])
```

```python
y=d['Purchased']
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x=s.fit_transform(x)
```

```python
x
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

## Stochastic GD

```python
model1 = Sequential()
model1.add(Dense(3, activation='relu', input_dim=2))
model1.add(Dense(3, activation='relu'))
model1.add(Dense(3, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model1.fit(x,y,epochs=50,validation_split=0.2,batch_size=1)
```

## Batch GD

```python
model2 = Sequential()
model2.add(Dense(3, activation='relu', input_dim=2))
model2.add(Dense(3, activation='relu'))
model2.add(Dense(3, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model2.fit(x,y,batch_size=200,validation_split=0.2,epochs=50)
```
